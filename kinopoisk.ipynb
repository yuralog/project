```python
def links_from_page(link):
    p = requests_my(link)
    soap = bs4.BeautifulSoup(p.text, "html.parser")
    soap = soap.select('.tenItems')[0].select('div')
    links = []
    names = []
    for i in range(0, len(soap)):
        try:
            names.append(soap[i].select('a')[0].select('img')[0]['alt'])
            links.append('https://www.kinopoisk.ru' + soap[i].select('a')[0]['href'])
            
            
        except:
            continue
    df = pd.DataFrame(names, links).reset_index()
    df = df.drop_duplicates()
    return df
```

```python
def one_film(link):
    p = requests_my(link)
    soap = bs4.BeautifulSoup(p.text, "html.parser")
    soap = soap.select('table.info')[0].select('td') 
    header = []
    s = []
    for i in range(0, len(soap),2):
        
        header.append(str(soap[i]).split('</td>')[0].split('>')[1])
        s.append(soap[i+1])
    df = pd.DataFrame(columns = header)
    df.loc[0] = s
    return df
```
